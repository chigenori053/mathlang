ã“ã®ãƒãƒ¼ãƒˆãƒ–ãƒƒã‚¯ã§ã¯ã€MathLang Core Extended ã®ä¸»è¦æ©Ÿèƒ½ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

## æ©Ÿèƒ½ä¸€è¦§
1. **ComputationEngine** - å¼ã®è¨ˆç®—ã€ç°¡ç´„åŒ–ã€å±•é–‹ã€å› æ•°åˆ†è§£
2. **ValidationEngine** - ç­”ãˆã®æ¤œè¨¼ï¼ˆè¨˜å·çš„ç­‰ä¾¡æ€§ã€å®Œå…¨ä¸€è‡´ã€æ¨™æº–å½¢å¼ï¼‰
3. **HintEngine** - ãƒ’ãƒ³ãƒˆç”Ÿæˆï¼ˆãƒ‘ã‚¿ãƒ¼ãƒ³ãƒãƒƒãƒãƒ³ã‚°ã€ç¬¦å·ã‚¨ãƒ©ãƒ¼ã€å®šæ•°ã‚ªãƒ•ã‚»ãƒƒãƒˆï¼‰
4. **CoreRuntime** - å•é¡Œè§£æ±ºã®å®Œå…¨ãªãƒ©ã‚¤ãƒ•ã‚µã‚¤ã‚¯ãƒ«

```python
# å¿…è¦ãªãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
from core.symbolic_engine import SymbolicEngine
from core.computation_engine import ComputationEngine
from core.validation_engine import ValidationEngine
from core.hint_engine import HintEngine
from core.core_runtime import CoreRuntime
from core.exercise_spec import ExerciseSpec
```

## 1. ComputationEngine ã®ãƒ†ã‚¹ãƒˆ

å¼ã®è¨ˆç®—ã€ç°¡ç´„åŒ–ã€å±•é–‹ã€å› æ•°åˆ†è§£ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

```python
# ã‚¨ãƒ³ã‚¸ãƒ³ã®åˆæœŸåŒ–
symbolic_engine = SymbolicEngine()
computation_engine = ComputationEngine(symbolic_engine)

print("=" * 60)
print("ComputationEngine ã®ãƒ†ã‚¹ãƒˆ")
print("=" * 60)

# ç°¡ç´„åŒ–
print("\nã€ç°¡ç´„åŒ–ã€‘")
expr1 = "x + x + x"
result1 = computation_engine.simplify(expr1)
print(f"{expr1} â†’ {result1}")

expr2 = "2*x + 3*x"
result2 = computation_engine.simplify(expr2)
print(f"{expr2} â†’ {result2}")

# å±•é–‹
print("\nã€å±•é–‹ã€‘")
expr3 = "(x + 1)**2"
result3 = computation_engine.expand(expr3)
print(f"{expr3} â†’ {result3}")

expr4 = "(x - 3)*(x + 3)"
result4 = computation_engine.expand(expr4)
print(f"{expr4} â†’ {result4}")

# å› æ•°åˆ†è§£
print("\nã€å› æ•°åˆ†è§£ã€‘")
expr5 = "x**2 - 9"
result5 = computation_engine.factor(expr5)
print(f"{expr5} â†’ {result5}")

expr6 = "x**2 + 2*x + 1"
result6 = computation_engine.factor(expr6)
print(f"{expr6} â†’ {result6}")

# æ•°å€¤è©•ä¾¡
print("\nã€æ•°å€¤è©•ä¾¡ã€‘")
computation_engine.bind("x", 5)
expr7 = "x**2 + 2*x + 1"
result7 = computation_engine.numeric_eval(expr7)
print(f"{expr7} (x=5) â†’ {result7}")
```

## 2. ValidationEngine ã®ãƒ†ã‚¹ãƒˆ

ç­”ãˆã®æ¤œè¨¼æ©Ÿèƒ½ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

```python
# ã‚¨ãƒ³ã‚¸ãƒ³ã®åˆæœŸåŒ–
validation_engine = ValidationEngine(computation_engine)

print("=" * 60)
print("ValidationEngine ã®ãƒ†ã‚¹ãƒˆ")
print("=" * 60)

# è¨˜å·çš„ç­‰ä¾¡æ€§ã®ãƒ†ã‚¹ãƒˆ
print("\nã€è¨˜å·çš„ç­‰ä¾¡æ€§ã€‘")
spec1 = ExerciseSpec(
    id="test1",
    target_expression="x**2 + 2*x + 1",
    validation_mode="symbolic_equiv"
)

# æ­£è§£ã®å ´åˆ
result1 = validation_engine.check_answer("(x + 1)**2", spec1)
print(f"ç­”ãˆ: (x + 1)**2")
print(f"çµæœ: {'âœ“ æ­£è§£' if result1.is_correct else 'âœ— ä¸æ­£è§£'}")
print(f"ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸: {result1.message}")

# ä¸æ­£è§£ã®å ´åˆ
result2 = validation_engine.check_answer("x**2 + x + 1", spec1)
print(f"\nç­”ãˆ: x**2 + x + 1")
print(f"çµæœ: {'âœ“ æ­£è§£' if result2.is_correct else 'âœ— ä¸æ­£è§£'}")
print(f"ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸: {result2.message}")

# æ¨™æº–å½¢å¼ã®ãƒ†ã‚¹ãƒˆ
print("\nã€æ¨™æº–å½¢å¼ã€‘")
spec2 = ExerciseSpec(
    id="test2",
    target_expression="x**2 - 9",
    validation_mode="canonical_form",
    canonical_form="x**2 - 9"
)

result3 = validation_engine.check_answer("x**2 - 9", spec2)
print(f"ç­”ãˆ: x**2 - 9")
print(f"çµæœ: {'âœ“ æ­£è§£' if result3.is_correct else 'âœ— ä¸æ­£è§£'}")
print(f"ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸: {result3.message}")
```

## 3. HintEngine ã®ãƒ†ã‚¹ãƒˆ

ãƒ’ãƒ³ãƒˆç”Ÿæˆæ©Ÿèƒ½ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

```python
# ã‚¨ãƒ³ã‚¸ãƒ³ã®åˆæœŸåŒ–
hint_engine = HintEngine(computation_engine)

print("=" * 60)
print("HintEngine ã®ãƒ†ã‚¹ãƒˆ")
print("=" * 60)

# ãƒ‘ã‚¿ãƒ¼ãƒ³ãƒãƒƒãƒãƒ³ã‚°
print("\nã€ãƒ‘ã‚¿ãƒ¼ãƒ³ãƒãƒƒãƒãƒ³ã‚°ã€‘")
spec3 = ExerciseSpec(
    id="test3",
    target_expression="x**2 + 2*x + 1",
    hint_rules={
        "x**2 + 1": "ä¸­é–“é …ã‚’å¿˜ã‚Œã¦ã„ã¾ã›ã‚“ã‹ï¼Ÿ",
        "x**2 + 2*x": "å®šæ•°é …ã‚’å¿˜ã‚Œã¦ã„ã¾ã›ã‚“ã‹ï¼Ÿ"
    }
)

hint1 = hint_engine.generate_hint_for_spec("x**2 + 1", spec3)
print(f"ç­”ãˆ: x**2 + 1")
print(f"ãƒ’ãƒ³ãƒˆ: {hint1.message}")
print(f"ã‚¿ã‚¤ãƒ—: {hint1.hint_type}")

# ç¬¦å·ã‚¨ãƒ©ãƒ¼ã®æ¤œå‡º
print("\nã€ç¬¦å·ã‚¨ãƒ©ãƒ¼ã®æ¤œå‡ºã€‘")
spec4 = ExerciseSpec(
    id="test4",
    target_expression="x + 5"
)

hint2 = hint_engine.generate_hint_for_spec("-x - 5", spec4)
print(f"ç­”ãˆ: -x - 5")
print(f"ãƒ’ãƒ³ãƒˆ: {hint2.message}")
print(f"ã‚¿ã‚¤ãƒ—: {hint2.hint_type}")

# å®šæ•°ã‚ªãƒ•ã‚»ãƒƒãƒˆã®æ¤œå‡º
print("\nã€å®šæ•°ã‚ªãƒ•ã‚»ãƒƒãƒˆã®æ¤œå‡ºã€‘")
hint3 = hint_engine.generate_hint_for_spec("x + 7", spec4)
print(f"ç­”ãˆ: x + 7")
print(f"ãƒ’ãƒ³ãƒˆ: {hint3.message}")
print(f"ã‚¿ã‚¤ãƒ—: {hint3.hint_type}")
print(f"ã‚ªãƒ•ã‚»ãƒƒãƒˆ: {hint3.details.get('offset', 'N/A')}")
```

## 4. CoreRuntime ã®ãƒ†ã‚¹ãƒˆ

å•é¡Œè§£æ±ºã®å®Œå…¨ãªãƒ©ã‚¤ãƒ•ã‚µã‚¤ã‚¯ãƒ«ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

```python
# CoreRuntimeã®åˆæœŸåŒ–
runtime = CoreRuntime(
    computation_engine=computation_engine,
    validation_engine=validation_engine,
    hint_engine=hint_engine
)

print("=" * 60)
print("CoreRuntime ã®ãƒ†ã‚¹ãƒˆ")
print("=" * 60)

# å•é¡Œã®è¨­å®š
print("\nã€å•é¡Œã€‘ x + x ã‚’ç°¡ç´„åŒ–ã—ã¦ãã ã•ã„")
runtime.set("x + x")

# æ­£ã—ã„ã‚¹ãƒ†ãƒƒãƒ—
print("\nã€ã‚¹ãƒ†ãƒƒãƒ—1ã€‘ 2*x")
result1 = runtime.check_step("2*x")
print(f"çµæœ: {'âœ“ æ­£ã—ã„' if result1['valid'] else 'âœ— é–“é•ã„'}")
if not result1['valid'] and 'hint' in result1['details']:
    print(f"ãƒ’ãƒ³ãƒˆ: {result1['details']['hint']['message']}")

# é–“é•ã£ãŸã‚¹ãƒ†ãƒƒãƒ—ï¼ˆç¬¦å·ã‚¨ãƒ©ãƒ¼ï¼‰
print("\nã€é–“é•ã£ãŸä¾‹ã€‘ -2*x")
runtime.set("x + x")  # ãƒªã‚»ãƒƒãƒˆ
result2 = runtime.check_step("-2*x")
print(f"çµæœ: {'âœ“ æ­£ã—ã„' if result2['valid'] else 'âœ— é–“é•ã„'}")
if not result2['valid'] and 'hint' in result2['details']:
    print(f"ãƒ’ãƒ³ãƒˆ: {result2['details']['hint']['message']}")
    print(f"ãƒ’ãƒ³ãƒˆã‚¿ã‚¤ãƒ—: {result2['details']['hint']['type']}")

# é–“é•ã£ãŸã‚¹ãƒ†ãƒƒãƒ—ï¼ˆå®šæ•°ã‚ªãƒ•ã‚»ãƒƒãƒˆï¼‰
print("\nã€é–“é•ã£ãŸä¾‹ã€‘ 2*x + 1")
runtime.set("x + x")  # ãƒªã‚»ãƒƒãƒˆ
result3 = runtime.check_step("2*x + 1")
print(f"çµæœ: {'âœ“ æ­£ã—ã„' if result3['valid'] else 'âœ— é–“é•ã„'}")
if not result3['valid'] and 'hint' in result3['details']:
    print(f"ãƒ’ãƒ³ãƒˆ: {result3['details']['hint']['message']}")
    print(f"ãƒ’ãƒ³ãƒˆã‚¿ã‚¤ãƒ—: {result3['details']['hint']['type']}")
```

## 5. ExerciseSpec ã‚’ä½¿ã£ãŸå®Œå…¨ãªãƒ†ã‚¹ãƒˆ

å•é¡Œä»•æ§˜ã‚’ä½¿ã£ãŸå®Œå…¨ãªãƒ¯ãƒ¼ã‚¯ãƒ•ãƒ­ãƒ¼ã‚’ãƒ†ã‚¹ãƒˆã—ã¾ã™ã€‚

```python
# å•é¡Œä»•æ§˜ã®ä½œæˆ
exercise = ExerciseSpec(
    id="quadratic_expansion",
    target_expression="x**2 + 2*x + 1",
    validation_mode="symbolic_equiv",
    hint_rules={
        "x**2 + 1": "ä¸­é–“é … 2*x ã‚’å¿˜ã‚Œã¦ã„ã¾ã›ã‚“ã‹ï¼Ÿ",
        "x**2 + x + 1": "ä¸­é–“é …ã®ä¿‚æ•°ã¯ 2 ã§ã‚ã‚‹ã¹ãã§ã™ã€‚"
    }
)

# CoreRuntimeã®åˆæœŸåŒ–ï¼ˆå•é¡Œä»•æ§˜ä»˜ãï¼‰
runtime_with_spec = CoreRuntime(
    computation_engine=computation_engine,
    validation_engine=validation_engine,
    hint_engine=hint_engine,
    exercise_spec=exercise
)

print("=" * 60)
print("ExerciseSpec ã‚’ä½¿ã£ãŸå®Œå…¨ãªãƒ†ã‚¹ãƒˆ")
print("=" * 60)
print(f"\nã€å•é¡Œã€‘ (x + 1)^2 ã‚’å±•é–‹ã—ã¦ãã ã•ã„")
print(f"æ­£è§£: {exercise.target_expression}")

# å­¦ç”Ÿã®å›ç­”ä¾‹1: ä¸­é–“é …ã‚’å¿˜ã‚ŒãŸ
print("\nã€å›ç­”1ã€‘ x**2 + 1")
result1 = runtime_with_spec.finalize("x**2 + 1")
print(f"çµæœ: {'âœ“ æ­£è§£' if result1['valid'] else 'âœ— ä¸æ­£è§£'}")
if not result1['valid'] and 'hint' in result1['details']:
    print(f"ğŸ’¡ ãƒ’ãƒ³ãƒˆ: {result1['details']['hint']['message']}")

# å­¦ç”Ÿã®å›ç­”ä¾‹2: ä¿‚æ•°ãŒé–“é•ã£ã¦ã„ã‚‹
print("\nã€å›ç­”2ã€‘ x**2 + x + 1")
result2 = runtime_with_spec.finalize("x**2 + x + 1")
print(f"çµæœ: {'âœ“ æ­£è§£' if result2['valid'] else 'âœ— ä¸æ­£è§£'}")
if not result2['valid'] and 'hint' in result2['details']:
    print(f"ğŸ’¡ ãƒ’ãƒ³ãƒˆ: {result2['details']['hint']['message']}")

# å­¦ç”Ÿã®å›ç­”ä¾‹3: æ­£è§£
print("\nã€å›ç­”3ã€‘ x**2 + 2*x + 1")
result3 = runtime_with_spec.finalize("x**2 + 2*x + 1")
print(f"çµæœ: {'âœ“ æ­£è§£' if result3['valid'] else 'âœ— ä¸æ­£è§£'}")
print(f"ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸: {result3['details'].get('message', 'å®Œç’§ã§ã™ï¼')}")
```
